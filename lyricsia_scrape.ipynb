```python
# Convert source.content to a beautifulsoup object 
# beautifulsoup can parse (extract specific information) HTML code

soup = bs.BeautifulSoup(source1.content, features='html.parser') 
# we pass in the source content
# features specifies what type of code we are parsing, 
# here 'html.parser' specifies that we want beautiful soup to parse HTML code
```

```python
navigation_bar = soup.find('form')
print(navigation_bar)
```

```python
# These are the linked subpages in the navigation bar
nav_bar = navigation_bar.text
print(nav_bar)
```

```python
link_list_year =[]
year=nav_bar.split("\n")
for y in year[3:-2]: #this index was to get only the year value in our list
    link = "https://www.lyricsia.com/hindi-songs/"+ y
    link_list_year.append(link)
```

```python
print(link_list_year[0])
```

```python
def get_lyrics(row):
    link = row.find('a')
    url = 'https://www.lyricsia.com'+link.get('href')
    song = requests.get(url)
    song_content = bs.BeautifulSoup(song.content,features='html.parser')
    lyrics = song_content.find('pre').text
    return lyrics
```

```python
df=pd.DataFrame(columns=["Song","Movie","Artist","Lyrics","Year"])
```

```python
for i in link_list_year:
    source = requests.get(i)
    soup = bs.BeautifulSoup(source.content, features='html.parser') 
    class_year = soup.find(class_='table table-striped')
    rows = class_year.find_all('tr')
    table_result = list()
    for row in rows[1:]:
        row_cells = row.find_all('td') # find all table data
        lyrics = get_lyrics(row_cells[1])
        row_entries = [cell.text for cell in row_cells[1:4]]
        #print(row_entries) 
        df=df.append({"Song":row_entries[0],"Movie":row_entries[1],
                      "Artist":row_entries[2],"Lyrics":lyrics,
                     "Year":i[-4:]},ignore_index=True)
        #table_result.append(row_entries)
        # get all the table data into a list
        break #(remove this while doing the entire thing)
```
