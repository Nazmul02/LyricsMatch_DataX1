```python
import numpy as np
dim = 1024
```

```python
hindi = np.fromfile("data/embeddings/pratik_hindi_embeddings.raw", dtype=np.float32, count=-1)                                                                          
hindi.resize(hindi.shape[0] // dim, dim)
```

```python
english = np.fromfile("data/embeddings/english_embeddings.raw", dtype=np.float32, count=-1)                                                                          
english.resize(english.shape[0] // dim, dim)
```

```python
d=1024
import faiss                   # make faiss available
index = faiss.IndexFlatL2(d)   # build the index
print(index.is_trained)
index.add(hindi)                  # add vectors to the index
print(index.ntotal)
```

```python
k = 4                          # we want to see 4 nearest neighbors
D, I = index.search(english[5:7], k) 
print(I)
print(D)
```
